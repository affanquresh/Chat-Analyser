```python
f=open("aiml_group.txt","r",encoding="utf-8")
```

```python
data=f.read()
```

```python
import re
# Regex pattern to extract timestamps and messages
pattern = r"\[(\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}:\d{2}\s?[APM]{2})\] (.+)"

# Extract timestamp and message pairs
matches = re.findall(pattern, data)

# Separate into two lists
dates, messages = zip(*matches)  # Unzipping into two separate variables

# Convert tuples to lists
dates = list(dates)
messages = list(messages)

# Print the extracted values
print("Dates:", dates)
print("Messages:", messages)
```

```python
import pandas as pd

# Convert dates to pandas datetime format
dates = pd.to_datetime(dates, format="%d/%m/%y, %I:%M:%S %p")

# Print converted dates
print(dates)
```

```python
import pandas as pd

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'User_Message': messages})

# Display the DataFrame
df.head()
```

```python
df.shape
```

```python
# Creating a new column 'User' with default value 'Group_Message'
df['User'] = df['User_Message'].apply(lambda x: x.split(':', 1)[0] if ':' in x else 'Group_Message')

# Creating a new column 'Message' that contains the actual message
df['Message'] = df['User_Message'].apply(lambda x: x.split(':', 1)[1].strip() if ':' in x else x)

# Dropping the old 'User_Message' column
df.drop(columns=['User_Message'], inplace=True)

# Display the updated DataFrame
df.head()
```

```python
media_keywords = ["image omitted", "video omitted", "GIF omitted", 
                  "sticker omitted", "document omitted", "audio omitted"]

# Convert 'Message' column to lowercase for case-insensitive comparison
df['Message'] = df['Message'].astype(str).str.lower().str.strip()

# Filter rows where 'Message' exactly matches any media keyword
media_df = df[df['Message'].isin(media_keywords)]

# Count the total media messages
total_media_shared = len(media_df)

# Display results
print("Total media messages shared:", total_media_shared)
 
```
